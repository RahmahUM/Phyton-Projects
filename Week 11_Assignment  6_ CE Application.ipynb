## Solving a basic 1D heat conduction problem with a constant thermal conductivity.

```python
import numpy as np
import matplotlib.pyplot as plt

def heat_conduction(length, temperature_left, temperature_right, thermal_conductivity, num_points):
    """
    Solves 1D heat conduction problem using finite difference method.

    Parameters:
    - length: Length of the material (m)
    - temperature_left: Temperature at the left boundary (°C)
    - temperature_right: Temperature at the right boundary (°C)
    - thermal_conductivity: Thermal conductivity of the material (W/(m*K))
    - num_points: Number of points/grid cells for discretization

    Returns:
    - temperatures: Array of temperatures along the material
    """

    dx = length / (num_points - 1)
    temperatures = np.zeros(num_points)
    
    # Boundary conditions
    temperatures[0] = temperature_left
    temperatures[-1] = temperature_right

    for _ in range(1000):  # Iterative solution
        for i in range(1, num_points - 1):
            temperatures[i] = 0.5 * (temperatures[i-1] + temperatures[i+1])

    return temperatures

# Example usage
length = 0.1  # meters
temperature_left = 100.0  # °C
temperature_right = 50.0  # °C
thermal_conductivity = 200.0  # W/(m*K)
num_points = 11

temperatures = heat_conduction(length, temperature_left, temperature_right, thermal_conductivity, num_points)

# Plotting the results
x_values = np.linspace(0, length, num_points)
plt.plot(x_values, temperatures, marker='o')
plt.xlabel('Position (m)')
plt.ylabel('Temperature (°C)')
plt.title('1D Heat Conduction')
plt.grid(True)
plt.show()
```
